```python
zf2 = zipfile.ZipFile('dataset/combined_data.zip') 
data = pd.read_csv(zf2.open('cobined_data.csv'))
```

```python
data.head()
```

```python
data = data.drop(['Unnamed: 0'],axis=1)
```

```python
data.target.value_counts().plot(kind='pie',autopct='%1.0f%%')
```

```python
data.groupby(['subject']).target.value_counts().plot.bar()
```

```python
data['clean_title'] = data['title'].apply(lambda x: " ".join(re.findall(r'[A-Za-z]{3,}', str(x).lower())))
data['clean_text'] = data['text'].apply(lambda x: " ".join(re.findall(r'[A-Za-z]{3,}', str(x).lower())))
```

```python
data = data.drop(['title','text'],axis=1)
```

```python
data['target'] = data['target'].replace(to_replace ="Fake",value = 0)
data['target'] = data['target'].replace(to_replace ="True",value = 1)
```

```python
data.head()
```

```python
count_vectorizer = CountVectorizer()
count_vectorizer.fit_transform(data.clean_text)
freq_term_matrix = count_vectorizer.transform(data['clean_text'])
tfidf = TfidfTransformer(norm = "l2")
tfidf.fit(freq_term_matrix)
tf_idf_matrix = tfidf.fit_transform(freq_term_matrix)
```

```python
train_text, test_text, train_labels, test_labels = train_test_split(tf_idf_matrix, data['target'], 
                                                                    random_state=2018, 
                                                                    test_size=0.3, 
                                                                    stratify=data['target'])
```

```python
count_vectorizer = CountVectorizer()
count_vectorizer.fit_transform(data.clean_title)
freq_term_matrix = count_vectorizer.transform(data.clean_title)
tfidf = TfidfTransformer(norm = "l2")
tfidf.fit(freq_term_matrix)
tf_idf_matrix = tfidf.fit_transform(freq_term_matrix)
```

```python
train_title, test_title, train_labels, test_labels = train_test_split(tf_idf_matrix, data['target'], 
                                                                    random_state=2018, 
                                                                    test_size=0.3, 
                                                                    stratify=data['target'])
```

# Logistic Regression 

```python
logreg = LogisticRegression()
logreg.fit(train_text, train_labels)
Accuracy = logreg.score(test_text, test_labels)
```

```python
print(Accuracy)
fig, ax = plot_confusion_matrix(conf_mat=confusion_matrix(test_labels, logreg.predict(test_text)),
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True)
```

```python
logreg = LogisticRegression()
logreg.fit(train_title, train_labels)
Accuracy = logreg.score(test_title, test_labels)
print(Accuracy)
fig, ax = plot_confusion_matrix(conf_mat=confusion_matrix(test_labels, logreg.predict(test_title)),
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True)
```

# Naive Bayes

```python
NB = MultinomialNB()
NB.fit(train_text, train_labels)
Accuracy_NB = NB.score(test_text, test_labels)
print(Accuracy_NB)
fig, ax = plot_confusion_matrix(conf_mat=confusion_matrix(test_labels, NB.predict(test_text)),
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True)
```

```python
NB = MultinomialNB()
NB.fit(train_title, train_labels)
Accuracy_NB = NB.score(test_title, test_labels)
print(Accuracy_NB)
fig, ax = plot_confusion_matrix(conf_mat=confusion_matrix(test_labels, NB.predict(test_title)),
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True)
```

# Passive Aggressive Classifier

```python
pac=PassiveAggressiveClassifier(max_iter=50)
pac.fit(train_text, train_labels)
y_pred=pac.predict(test_text)
score=accuracy_score(test_labels, y_pred)
print(f'Accuracy: {round(score*100,2)}%')
fig, ax = plot_confusion_matrix(conf_mat=confusion_matrix(test_labels, y_pred),
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True)
```

```python
pac=PassiveAggressiveClassifier(max_iter=50)
pac.fit(train_title, train_labels)
y_pred=pac.predict(test_title)
score=accuracy_score(test_labels, y_pred)
print(f'Accuracy: {round(score*100,2)}%')
fig, ax = plot_confusion_matrix(conf_mat=confusion_matrix(test_labels, y_pred),
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True)
```
